## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# suppress warnings
import warnings
warnings.filterwarnings('ignore')

#  Import and read the charity_data.csv.
df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
df.head()
```

```python
 # Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
col_drop = ["EIN", "NAME", "SPECIAL_CONSIDERATIONS"]

df.drop(columns = col_drop, inplace = True)
df.head()
```

```python
# Determine the number of unique values in each column.
columns = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
       'ORGANIZATION', 'STATUS', 'INCOME_AMT', 'ASK_AMT', 'IS_SUCCESSFUL']

for column in columns:
  print(f"""{column}:  {df[column].nunique()}""")
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_type_value_counts = df.APPLICATION_TYPE.value_counts()
application_types_to_replace = application_type_value_counts.loc[application_type_value_counts < 1000].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classification_value_counts = df.CLASSIFICATION.value_counts()
classifications_to_replace = classification_value_counts.loc[classification_value_counts < 1000].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
df['CLASSIFICATION'].value_counts()
```

```python
df.ASK_AMT.value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
ask_amt_value_counts = df.ASK_AMT.value_counts()
ask_amt_to_replace = ask_amt_value_counts.loc[ask_amt_value_counts < 1000].index.tolist()

# Replace in dataframe
for amt in ask_amt_to_replace:
    df['ASK_AMT'] = df['ASK_AMT'].replace(amt,"Other")

# Check to make sure replacement was successful
df['ASK_AMT'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df2 = pd.get_dummies(df, dtype = int)
df2.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = df2.drop(columns=["IS_SUCCESSFUL"])
y = df2.IS_SUCCESSFUL

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Model 5

nn5 = tf.keras.models.Sequential()

# First hidden layer
nn5.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=len(X.columns)))

# Output layer
nn5.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn5.summary()
```

```python
# Compile the model
nn5.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
train_model = nn5.fit(X_train, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn5.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn5.save("AlphabetSoupCharity_Model5.h5")
```
